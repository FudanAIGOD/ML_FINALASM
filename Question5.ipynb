```python
#导入所需要的库
import pandas as pd
from sklearn import tree
from sklearn.preprocessing import LabelEncoder
```

```python
#读取数据，并把题本中给的预测数据也写进来，假定buying为no,为便处理数据
df=pd.read_excel('Question5.xlsx')
df
```

```python
#为每一列序列化,就是将每种字符串转化为对应的数字。用数字代表类别
le = LabelEncoder()
for col in df.columns:                                            
    df[col] = le.fit_transform(df[col])
```

```python
#选取数据进行建模
X=df[['Age','Incoming','Student','Credit Rating']][:-1]
Y=df['Buying'][:-1]
```

```python
#建立决策树模型
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X, Y)
```

```python
#估计
preX=df[['Age','Incoming','Student','Credit Rating']][-1:]
result = clf.predict(preX) 
YDict={0:'no',1:'yes'}
print(YDict[result[0]])
```
